```python
import pandas as pd
import hashlib
import os
```

```python
df = pd.read_csv('checklist1.csv').set_index('names')
```

```python
for name in df.index:
    if (df.loc[name]['NConDISK']==0):
        path = '/nobackup/csjone15/pleiades_llc_recipes/python_cli_data_export/surf_extract/surf_fields/' +name + '.nc'
        if(os.path.isfile(path)):
            if(os.path.getsize(path)>10**6):
                df.loc[name]['NConDISK']=1
    if (df.loc[name]['JSONonDISK']==0):
        path_json = '/nobackup/csjone15/pleiades_llc_recipes/python_cli_data_export/surf_extract/surf_json/' +name + '.json'
        if(os.path.isfile(path_json)):
            df.loc[name]['JSONonDISK']=1
            with open(path_json, 'r') as fp:
                lines = fp.readlines(20)
                for row in lines:
                    if row.find("cnh-bucket-1") != -1:
                        df.loc[name]['jsonreplace']=1
```
